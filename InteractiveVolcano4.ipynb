```python
import plotly
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
init_notebook_mode(connected=True)
from IPython.display import Image 
im1 = Image(filename ="Legend1.png")
im2 = Image(filename ="Legend2.png")

import pandas as pd
import numpy as np
import plotly.graph_objects as go

df = pd.read_csv (r'PythonKey_CSA13vUntreated.csv') 
adjpval = df['adjpval']
opaq = df['opaq']
pointsize = df['pointsize']
hovinfo = df['hovinfo']
pointcolor = df['color']
Name = df['name']
avgexp = df['avgexp']
logfc = df['logfc']
adjpval = df['adjpval']

x=1*logfc
y=-1*np.log10(adjpval)

# Build figure
fig = go.Figure()

# Add scatter trace with medium sized markers
fig.add_trace(
    go.Scatter(
        mode='markers',
        x=x,
        y=y,
        text = Name.to_list(),
   
        hoverinfo=hovinfo.to_list(),
        
        marker=dict(
            color=pointcolor,
            size=pointsize,
            opacity=opaq,
            line=dict(
                color='black',
                width=0.15
            )
            
        ),
        showlegend=False
        
    )
)
fig.update_layout(
    title={
        'text': "Untreated E. coli vs 8ug/ml CSA13",
        'y':0.85,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'},
    xaxis_title="-log10(Adjusted P value)",
    yaxis_title="Log2 Fold Change")

fig.add_layout_image(
        dict(
            source="Legend1.png",
            xref="x",
            yref="y",
            x=5,
            y=3,
            sizex=3,
            sizey=3,
            opacity=1,
            layer="above")
)

fig.add_layout_image(
        dict(
            source="Legend2.png",
            xref="x",
            yref="y",
            x=-8.3,
            y=9,
            sizex=2,
            sizey=2,
            opacity=1,
            layer="above")
)
```
